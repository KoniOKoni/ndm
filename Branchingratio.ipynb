```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
me = 0.5e6
mmu = 105e6
p = 50e6
e = np.sqrt(4*np.pi/137)
UU = 0.01
GF = 1e-23
Gamma = GF**2 * mmu**5 / (192 * np.pi**3)
Gammatau = 2.27e-3
mtau = 1776e6

ye = 2.94e-6
ymu = 6.07e-4
ytau = 1.02e-2

ptau =880e6
```

```python
def eg(M, expsum):
    Isq = (M**6 * p**2)/(mmu**7) * e**2 * ye**2 * ymu**2 * UU**2 * expsum
    return Isq/Gamma

def eee(M, expsum):
    Isq = (M**8 * mmu**5)/(p**(12)) * e**4 * ye**2 * ymu**2 * UU**2 * expsum
    return Isq/Gamma

def egg(M, expsum):
    Isq = (M**2 * mmu**3)/(p**4) * e**4 * ye**2 * ymu**2 * UU**2 * expsum
    return Isq/Gamma

def tg(M, expsum):
    Isq = (M**6 * ptau**2)/(mtau**7) * e**2 * ye**2 * ytau**2 * UU**2 * expsum
    return Isq/Gammatau

def tee(M, expsum):
    Isq = (M**8 * mtau**5)/(ptau**(12)) * e**4 * ye**2 * ytau**2 * UU**2 * expsum
    return Isq/Gammatau

def tgg(M, expsum):
    Isq = (M**2 * mtau**3)/(ptau**4) * e**4 * ye**2 * ytau**2 * UU**2 * expsum
    return Isq/Gammatau

def tgmu(M, expsum):
    Isq = (M**6 * ptau**2)/(mtau**7) * e**2 * ytau**2 * ymu**2 * UU**2 * expsum
    return Isq/Gammatau

def teemu(M, expsum):
    Isq = (M**8 * mtau**5)/(ptau**(12)) * e**4 * ytau**2 * ymu**2 * UU**2 * expsum
    return Isq/Gammatau

def tggmu(M, expsum):
    Isq = (M**2 * mtau**3)/(ptau**4) * e**4 * ytau**2 * ymu**2 * UU**2 * expsum
    return Isq/Gammatau
```

```python
M = np.logspace(12, 20, 1000)
expsum = np.logspace(-80, 0, 1000)
M, expsum = np.meshgrid(M, expsum)
```

```python
eg(M, expsum)
```

```python
plt.contourf(M, expsum, eg(M, expsum), levels=[3.1e-13, 1e1000], colors='skyblue', alpha=0.3)
cs1 = plt.contour(M, expsum, eg(M, expsum), levels=[3.1e-13], colors='skyblue')
plt.text(1e13, 1e-35, r'$\mu \to e\gamma$')

plt.contourf(M, expsum, eee(M, expsum), levels=[1e-12, 1e1000], colors='orange', alpha=0.3)
cs2 = plt.contour(M, expsum, eee(M, expsum), levels=[1e-12], colors='orange')
plt.text(1e15, 1e-65, r'$\mu \to eee$')

plt.contourf(M, expsum, egg(M, expsum), levels=[7.2e-11, 1e1000], colors='purple', alpha=0.3)
cs3 = plt.contour(M, expsum, egg(M, expsum), levels=[7.2e-11], colors='purple')
plt.text(1e19, 1e-25, r'$\mu \to e\gamma\gamma$')

plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$M$ (eV)')
plt.ylabel(r'$\exp(-16\pi^2\sum\frac{1}{g^2})$')
plt.title(r'Muon CLFV constraints')
plt.savefig('BR_muon.pdf', dpi=300)
```

```python
plt.contourf(M, expsum, tg(M, expsum), levels=[3.3e-8, 1e1000], colors='red', alpha=0.3)
cs1 = plt.contour(M, expsum, tg(M, expsum), levels=[3.3e-8], colors='red')
plt.text(1e16, 1e-35, r'$\tau \to e\gamma$')

plt.contourf(M, expsum, tee(M, expsum), levels=[2.7e-8, 1e1000], colors='blue', alpha=0.3)
cs1 = plt.contour(M, expsum, tee(M, expsum), levels=[2.7e-8], colors='blue')
plt.text(1e16, 1e-56, r'$\tau \to eee$')

plt.contourf(M, expsum, tgg(M, expsum), levels=[2.5e-4, 1e1000], colors='green', alpha=0.3)
cs1 = plt.contour(M, expsum, tgg(M, expsum), levels=[2.5e-4], colors='green')
plt.text(1e16, 1e-8, r'$\tau \to e\gamma\gamma$')

plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$M$ (eV)')
plt.ylabel(r'$\exp(-16\pi^2\sum\frac{1}{g^2})$')
plt.title(r'Tau to e CLFV constraints')
plt.savefig('BR_tau_e.pdf', dpi=300)
```

```python
plt.contourf(M, expsum, tgmu(M, expsum), levels=[4.2e-8, 1e1000], colors='red', alpha=0.3)
cs1 = plt.contour(M, expsum, tgmu(M, expsum), levels=[4.2e-8], colors='red')
plt.text(1e16, 1e-40, r'$\tau \to \mu\gamma$')

plt.contourf(M, expsum, teemu(M, expsum), levels=[1.9e-8, 1e1000], colors='blue', alpha=0.3)
cs1 = plt.contour(M, expsum, teemu(M, expsum), levels=[1.9e-8], colors='blue')
plt.text(1e16, 1e-58, r'$\tau \to \mu\mu\mu$')

plt.contourf(M, expsum, tggmu(M, expsum), levels=[5.8e-4, 1e1000], colors='green', alpha=0.3)
cs1 = plt.contour(M, expsum, tggmu(M, expsum), levels=[5.8e-4], colors='green')
plt.text(1e16, 1e-8, r'$\tau \to \mu\gamma\gamma$')

plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$M$ (eV)')
plt.ylabel(r'$\exp(-16\pi^2\sum\frac{1}{g^2})$')
plt.title(r'Tau to mu CLFV constraints')
plt.savefig('BR_tau_mu.pdf', dpi=300)
```
