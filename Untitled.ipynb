```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
g0 = np.linspace(0.6, 0.7, 100)
Lambda = 1e25
v = np.logspace(13, 19, 100)

g0, v = np.meshgrid(g0, v)
gv = np.exp(-8*np.pi**2 * (np.power(g0, -2) + 35/(6*8*np.pi**2) * np.log(v/Lambda)))

plt.contourf(g0, v, gv,[1e-10, 1e-3, 1e-1], vmax = 1)
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$g_0$')
plt.ylabel(r'$v$')
```
